```python
import pandas as pd
import glob
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
directory_path = '../data/raw/2016-2021_player_stats/'
```

### Passing Stats csv aggregation

```python
# Getting all passing stats files for 2016-2021
master_df = pd.DataFrame()
files = glob.glob(directory_path+'*player_passing.csv')
for i in range(0, len(files)):
    f = files[i]
    year = int(files[i].split('/')[-1].split('_')[0])

    # read csv file for passing statistics
    df = pd.read_csv(f)
    df.insert(0, column='year', value=year)
    df.insert(0, column='stats_type', value='passing')
    # append to master_df
    master_df = pd.concat([master_df, df], ignore_index=True)

    # print the location and filename 
    print('Location:', f) 
    print('File Name:', f.split("\\")[-1]) 
      
    # print the content 
    print(year)
    print(len(df))
    print('Content:') 
    display(master_df) 
    print() 
```

```python
master_df.info()
```

```python
# no blatant na values in df. 
master_df.isna().sum()
```

```python
# No duplicated rows
master_df.loc[master_df.duplicated() == True]
```

```python
# Add column to indicate passer as superbowl winner from that year from cleaned, aggregated data file
superbowl_winning_roster = pd.read_csv('../data/cleaned/all_superbowl_winning_rosters.csv')

# Create key to match players from superbowl roster file and passer file
master_df['year_player'] = master_df['year'].astype(str) + '_' +  master_df['Player']
superbowl_winning_roster['year_player'] = superbowl_winning_roster['year'].astype(str) + '_' + superbowl_winning_roster['player']

# Creating column to indicate if player was part of superbowl winning team in year based on year_player key
master_df['won_super_bowl'] = np.where(master_df['year_player'].isin(superbowl_winning_roster['year_player']), 1, 0)
```

```python
# Passing attempts include a lot of few-time passers in a season. Looking for the distribution of these to find 'genuine' passers.
master_df_test_qbs = master_df.loc[master_df['Att'] > 20]
```

```python
# look for outliers in low frequency range of passing attempts (trick plays)
sns.histplot(data=master_df, x='Att', hue='won_super_bowl')
plt.show()
```

```python
# look for outliers in low frequency range of passing attempts (trick plays)
sns.histplot(data=master_df_test_qbs, x='Att', hue='won_super_bowl')
plt.show()
```

```python
master_df_test_qbs.head()
```

```python
master_df_test_qbs.to_csv("../data/cleaned/2016-2021_passing_stats.csv", index=False)
```
